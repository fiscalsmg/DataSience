
# Leer .csv de multiples URLs con pandas
link: root + season + lenguage

https://www.football-data.co.uk/mmz4281/2122/E0.csv

```python
# Estructura de un link
"https://www.football-data.co.uk/mmz4281/" + "2122" + "/" + "E0" + ".csv"
```

```python
# Crear una variable root
root = "https://www.football-data.co.uk/mmz4281/"
```

## 2.1 Multiples ligas

- https://www.football-data.co.uk/mmz4281/2122/E0.csv
- https://www.football-data.co.uk/mmz4281/2122/E1.csv
- https://www.football-data.co.uk/mmz4281/2122/E2.csv
- https://www.football-data.co.uk/mmz4281/2122/E3.csv
- https://www.football-data.co.uk/mmz4281/2122/EC.csv

```python
# Crear una lista de ligas

leagues = ["E0", "E1", "E2", "E3", "EC"]

frames = []

# hacer loop en la lista de ligas, leer multiples csv y adjuntarlo en una lista
for league in leagues:
    df = pd.read_csv(root + "2122" + "/" + league + ".csv", encoding="unicode_escape")
    # Guardar cada dataframe por iteracion 
    frames.append(df)

# frames

```

```python
# Tamaño de los frames
len(frames)
```

```python
# mostrar 1er, 2do y 3er elemento 
frames[4]
```

## 2.2 multiple temporadas

```python
# Hacer loop a un rango de temporadas e insertar una nueva columna 

leagues = ["E0", "E2", "E3"]

frames = []

# multiples ligas con diferentes temporadas 
for league in leagues:
    for season in range(15, 21):
        df = pd.read_csv(root + str(season)+ str(season + 1) + "/" + league + ".csv")
        # Crear columna que identifique a que temporada pertenece cada archivo csv
        # df["season"] = season # inserta pero en la ultima posicion
        df.insert(1, "season", season)
        # Guardar cada dataframe por iteracion 
        frames.append(df)

```

```python
# Total de frames (6 temporadas x 3 ligas)
len(frames)
```

```python
# Mostrar 1er y ultimo elemento (Columna season fue creada)
frames[17]
```

# 3. Organizar Data en un Diccionario

```python
# Crear un diccionario con nombre original de liga como llave
dict_countries = {'Spanish La Liga':'SP1',
                  'Spanish Segunda Division':'SP2',
                  'German Bundesliga':'D1',
                  'English Premier League':'E0', 
                  'English League 1':'E2', 
                  'English League 2':'E3',
                 }
```

```python
# Obtener un elemento del diccionario
dict_countries["Spanish La Liga"]
```

```python
# Hacer loop sobre las llaves del diccionario

dict_historic_data = {}


for league in dict_countries:
    frames = [] # por cada liga a la cual se itera
    for season in range(15, 21):
        df = pd.read_csv(root + str(season)+ str(season + 1) + "/" + dict_countries[league] + ".csv")
        df.insert(1, "season", season)
        frames.append(df)
        
    df_concat = pd.concat(frames)
    dict_historic_data[league] = df_concat
```

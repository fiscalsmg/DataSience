# 1. Agregar nueva columna 
Cuando es conveniente agregar cada metodo

## 1.1 assign()
¿Cuando usar assign?

- Agregar multiples columnasen una sola linea de codigo
- Cuando se necesita sobreescribir los valores de una columna exitente (buena practica)

Devuelve un nuevo objeto (copia) con todas las columnas originales ademas de las nuevas


```python
import numpy as np
```

```python
# Crear numeros aleatorios para nuestra nueva columna "score"
score_1 = np.random.randint(1,100, size=1000)
score_2 = np.random.randint(1,100, size=1000)
```

```python
# Crear una serie usando los numeros aleatorios 
serie_1 = pd.Series(score_1, index=np.arange(0, 1000))
serie_2 = pd.Series(score_2, index=np.arange(0, 1000))
```

```python
serie_1
```

```python
# Usando el metodo assing() para agregar multiples columnas 
df_exams.assign(score1 = score_1, score2 = score_2) # Asi solo es una copia
```

```python
# Aqui se lo asignamos al dataFrame
df_exams = df_exams.assign(score1 = score_1, score2 = score_2)
```

```python
# Mostrar dataFrame
df_exams
```

## 2. insert()

Insertar una nueva columna en una posicion especifica index

```python
# usar insert() para adicionar una columna en una posicion especifica 
df_exams.insert(1, "test", serie_1) # NO devuelve copia, solo actualiza el df
```

```python
# Mostrar dataFrame
df_exams
```

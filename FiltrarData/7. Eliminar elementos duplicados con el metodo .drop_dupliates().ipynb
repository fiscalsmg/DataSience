```python
import pandas as pd
```

```python
df_laptops = pd.read_csv("../Material/laptop_price.csv", encoding="latin-1")
df_laptops.head(3)
```

# 1. drop_duplicates()

```python
# Eliminar duplciados en 2 o mas columnas 
df_laptops.drop_duplicates(["Company"])[["Company", "Price_euros"]]
```

```python
# Ordenar DataFrame ascendente por compa√±ia (Company) y precio (price)
# mas barato primero y mas caro al ultimo 
df_laptops = df_laptops.sort_values(["Company", "Price_euros"]) # Ordenado ascendentemente
```

```python
# Mas barato: keep ="first"
df_laptops.drop_duplicates(["Company"], keep="first")[["Company", "Price_euros"]]
```

```python
# Mas caro: keep ="last"
df_laptops.drop_duplicates(["Company"], keep="last")[["Company", "Price_euros"]]
```

```python
# Argumentos
# inplace: Elimina duplicados en el dataFrame o devuelve copia
# ignore_index= True: El index resultadotendra etiqueta de 0, 1... renombra
df_laptops.drop_duplicates(["Company"], keep="last", inplace=True, ignore_index=True)
```

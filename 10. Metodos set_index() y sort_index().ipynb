# Crear index

set_index() - nos ayuda a fijar un index

sort_index() - nos ayuda a ordenar los index

```python
import numpy as np
import random
```

```python
# Crear valores no repetitivos para los index
new_index = np.arange(0,1000)
```

```python
# Barajear/desordenar los indices 
random.shuffle(new_index)
```

```python
new_index
```

```python
# Crear una nueva columna con los nuevos indices
df_exams["nuevo_index"] = new_index
df_exams
```

# 2. set_index()

```python
# Fijar columna nuevo_index como index

# CAMBIA el index UN INDEX NO PUEDE TENER REPETITIVO
df_exams.set_index("nuevo_index", inplace=True)
```

```python
df_exams
```

# 3. sort_index()

```python
# Ordenar dataFrame por nuevo_index
df_exams.sort_index()
```

```python
# Ordenar descendente por index
df_exams.sort_index(ascending=False, inplace=True)
```
